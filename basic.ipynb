```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d=sns.load_dataset('tips')
```

```python
d
```

```python
d['total_bill']
```

```python
plt.plot(d['total_bill'])
plt.show()
```

```python
sns.lineplot(d['total_bill'])
```

```python
sns.scatterplot(x='total_bill',y='tip',data=d)
```

```python
sns.histplot(d['total_bill'])
```

```python
sns.kdeplot(d['total_bill'])
```

```python
#relational plot
sns.relplot(x='total_bill',y='tip',data=d)
```

```python
d['time'].unique()
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',data=d)
```

```python
d['smoker'].unique()
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',style='smoker',data=d)
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',style='smoker',kind='scatter',data=d,row="time")
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',data=d,kind='scatter',column="smoker")
```

```python
d['day'].unique()
```

```python
sns.relplot(x='total_bill', y='tip', hue='time', kind='scatter', data=d, col='day')
```

```python
sns.relplot(x='total_bill',y='tip',data=d,)
plt.title("this is a scatter plot")
plt.xlabel("total_bill",fontsize=20,color='black')
plt.ylabel("tips",fontsize=20,color='black')
plt.show()
```

```python
#catplot
sns.catplot(x='day',y='total_bill',kind='strip',data=d)
```

```python
#catplot
sns.catplot(x='day',y='total_bill',kind='swarm',data=d)
```

```python
#catplot
sns.catplot(x='day',y='total_bill',kind='box',data=d)
```

```python
#catplot
sns.catplot(x='day',y='total_bill',kind='bar',data=d)
```

```python
#catplot
sns.catplot(x='day',y='total_bill',kind='violin',data=d)
```

```python
#catplot
sns.catplot(x='day',y='total_bill',kind='boxen',data=d)
```

```python
#countplot
plt.figure(figsize=(5,5))
sns.countplot(x='day',data=d)
```
